####Task6: Create a pyspark dataframe df1 reading the data from the above file using pyspark magic command %python

```python
%python
df1 = spark.read.csv("/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",header=True)
df1.head(5)
```

####Task7: Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command %python

```python
#As column name should not contain spaces or other special characters, we will rename the columns
df1_clean = df1.withColumnRenamed(
    "Mobile Operating System", "mobile_operating_system"
).withColumnRenamed(
    "Percent of Usage", "percent_of_usage"
)
```

```python
#used this cell to drop the table, in order to create with mentioned format
'''%sql
drop table if exists default.mobile_os_usage'''
```

```python
%python
df1_clean.write.saveAsTable("default.mobile_os_usage")
```
